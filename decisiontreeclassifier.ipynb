```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
d=pd.read_csv("/kaggle/input/car-data/cardata.csv")
d.dtypes
```

```python
d.head(4)
```

```python
import category_encoders as ce
ec=ce.OrdinalEncoder(cols=['Car_Name','Year','Selling_Price','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner'])
d= ec.fit_transform(d)
```

```python
d.head()
```

```python
from sklearn.model_selection import train_test_split
ektrain,ektest=train_test_split(d,test_size=0.3,random_state=2)
ektrain
```

```python
ektrain.reset_index(drop=True,inplace=True)
ektest.reset_index(drop=True,inplace=True)
ektrain
```

```python
 
xtrain=ektrain.drop(["Transmission"],axis=1)
ytrain=ektrain["Transmission"]
xtest=ektest.drop(["Transmission"],axis=1)
ytest=ektest["Transmission"]
xtrain
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(xtest)
print("accuracy score=",accuracy_score(ytest,y_pred)*100)
```

```python
from sklearn.metrics import classification_report
print("classification report",classification_report(ytest,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("confusion matrix",confusion_matrix(ytest,y_pred))
```
